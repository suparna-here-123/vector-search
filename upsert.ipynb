#### Rename recipe files to Rx_recipe_name

```python
import sys, os
R_ID = 1

parent = './recipes'
for child in os.listdir('./recipes') :
    new_name = f'R{R_ID}-' + child
    os.rename(os.path.join(parent, child), os.path.join(parent, new_name))
    R_ID += 1
```

#### Chunking every recipe and upserting into Pinecone 
1. Preprocess every recipe to remove newlines/carriage returns between paragraphs
2. Try different chunking methods :
    - Fixed (no.of lines-based)
    - Section-based (ingredients, method)


```python
# Currently making every 5 lines of the recipe 1 chunk
for file in os.listdir('./recipes') :
    with open(os.path.join(parent, file), 'r') as f :
        chunks = []
        curChunk = ''
        for lineNum, line in enumerate(f.readlines(), start = 1) :
            if line :                           # Line not empty
                if lineNum % 5 == 0 :
                    chunks.append(curChunk)
                    curChunk = ''
                else :
                    curChunk += ". " + line
        if curChunk :
            chunks.append(curChunk)

        print(chunks)
        break
```
